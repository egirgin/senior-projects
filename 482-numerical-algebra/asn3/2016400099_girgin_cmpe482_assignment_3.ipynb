```python
def model2(x_train, y_train, l2 = None):
    num_features = x_train.shape[1]
    try:
        thetas = np.linalg.inv(x_train.T.dot(x_train) + l2 * np.eye(num_features)).dot(x_train.T).dot(y_train)
        return thetas
    except Exception as e:
        print("Error: {}. Taking Pseudo-Inverse...".format(e))
        thetas = np.linalg.pinv(x_train.T.dot(x_train)).dot(x_train.T).dot(y_train)
        return thetas
```

```python
lambda1 = 0.00001
lambda2 = 100000
theta_1 = model2(x_train, y_train, lambda1)
theta_2 = model2(x_train, y_train, lambda2)
print("Theta 1: {}".format(theta_1))
print("Theta 2: {}".format(theta_2))
```

$\tilde\theta_{1}$ is closer to $\hat{\theta}$ because we used a very little regularization coefficient for $\tilde\theta_{1}$, which makes it very similar to original $\hat{\theta}$. On the other hand, $\tilde\theta_{2}$ has smaller weights compared to both $\hat{\theta}$ and $\tilde\theta_{1}$. This is because when higher regularization coefficient is used, the model is penalized more for bigger parameters.

### b. 

```python
lambdas = np.array([0.0, 0.001, 0.01, 0.1, 1.0])

train_rmse = []
test_rmse = []
for l2 in lambdas:
    model = model2(x_train, y_train, l2)
    
    pred_train = x_train.dot(model)
    train_error = rmse(pred_train, y_train)
    
    pred_test = x_test.dot(model)
    test_error = rmse(pred_test, y_test)
    
    train_rmse.append(train_error)
    test_rmse.append(test_error)
    print("Lambda: {} | Train RMSE: {} | Test RMSE: {}".format(l2, train_error, test_error))
    

plt.plot(lambdas, train_rmse, label="Train")
plt.plot(lambdas, test_rmse, label="Test")

plt.legend()
plt.show()
    
```

### d. 

By introducing _Lagrange Multipliers_ z (15-vector), we can convert our optimization problem to a _KKT Equations_:

$
\begin{bmatrix}
2A^{T}A & C^{T} \\
C & 0
\end{bmatrix}
\begin{bmatrix}
c \\
z
\end{bmatrix}
= 
\begin{bmatrix}
2A^{T}y \\
d
\end{bmatrix}
$, where the coefficient matrix is called _KKT Matrix_.

In order to find $c$, KKT Matrix has to be invertible and KKT Matrix is invertible if and only if
$ C$ has linearly independent rows and
$
\begin{bmatrix}
A \\
C
\end{bmatrix}
$
has linearly independent columns.

By rewriting equation
$
2A^{T}Ac + C^{T}z = 2A^{T}y \\
2A^{T}Ac + 0 + C^{T}z = 2A^{T}y \\
2A^{T}Ac + Cc + C^{T}z = 2A^{T}y \\
2A^{T}Ac + C^{T}Cc + C^{T}z = 2A^{T}y \\
2A^{T}Ac + 2C^{T}Cc + C^{T}z = 2A^{T}y \\
2(A^{T}A + C^{T}C)c + C^{T}z = 2A^{T}y \\
$
and 
$
Cc = d = 0
$

$
\begin{bmatrix}
A \\ 
C
\end{bmatrix}
= QR =
\begin{bmatrix}
Q_{1} \\
Q_{2}
\end{bmatrix} R
\rightarrow
A = Q_{1}R$ and $
C = Q_{2}R
$ and $ A^{T}A + C^{T}C = R^{T}R$

By plugging $Q$s and $R$s to the equations above, we obtain: 
$ 2R^{T}Rc + R^{T}Q_{2}^{T}z = 2R^{T}Q_{1}^{T}y $ and $Q_{2}Rc = d$

By multiplying $R^{-T}$ we get
$Rc = Q_{1}^{T}y - (1/2)Q_{2}^{T}z$ then substitute $Q_{2}Rc = d$ and get $Q_{2}Q_{2}^{T}z = 2Q_{2}Q_{1}^{T}b$ since d is zero matrix.

Also de-compose $Q_{2}^{T}$ as $Q_{2}^{T} =  \tilde{Q}\tilde{R}$ and put to the equation.

$\tilde{R}^{T}\tilde{R}z = 2\tilde{R}^{T}\tilde{Q}^{T}Q_{1}^{T}y$, which can be written as: $\tilde{R}z = 2\tilde{Q}^{T}Q_{1}^{T}y$

```python
Q, R = np.linalg.qr(np.vstack((A, C_matrix)))
Q_1 = Q[:150]
Q_2 = Q[150:]
Q_tilde, R_tilde = np.linalg.qr(Q_2.T)
R_tilde_inv = np.linalg.inv(R_tilde)
```

Since we know $\tilde{R}$ is invertible, we can calculate $z = \tilde{R}^{-1}2\tilde{Q}^{T}Q_{1}^{T}y$

```python
z = R_tilde_inv @ (2 * Q_tilde.T @ Q_1.T @ ys) 
```

Now put $z$ value to equation we obtained before:
$Rc = Q_{1}^{T}y - (1/2)Q_{2}^{T}z \rightarrow c = R^{-1}(Q_{1}^{T}y - (1/2)Q_{2}^{T}z)$

```python
coeffs = np.linalg.inv(R) @ (Q_1.T @ ys.reshape(-1,1) - 0.5 * Q_2.T.dot(z.reshape(-1,1)))
```

```python
plt.stem(coeffs)
plt.title("C Stem-Plot")
plt.show()
```

### f. 

KKT conditions: 
1) C has linearly independent rows, \
2) $\begin{bmatrix}
    A \\
    C
\end{bmatrix}$has linearly independent columns. \
Then equation $
\begin{bmatrix}
2A^{T}A & C^{T} \\
C & 0
\end{bmatrix}
\begin{bmatrix}
c \\
z
\end{bmatrix}
= 
\begin{bmatrix}
2A^{T}y \\
d
\end{bmatrix}
$ has a solution:
$
\begin{bmatrix}
c \\
z
\end{bmatrix}
=
\begin{bmatrix}
2A^{T}A & C^{T} \\
C & 0
\end{bmatrix}^{-1}
\begin{bmatrix}
2A^{T}y \\
d
\end{bmatrix}
$

```python
inv_coeffs_upper = np.hstack((2*A.T.dot(A), C_matrix.T))
inv_coeffs_lower = np.hstack((C_matrix, np.zeros((6,6))))
inv_coeffs = np.vstack((inv_coeffs_upper, inv_coeffs_lower))
inv_coeffs = np.linalg.inv(inv_coeffs)

mulptlier = np.hstack((2*A.T @ ys, d.reshape(-1,)))
```

```python
x_z = inv_coeffs @ mulptlier
x = x_z[:15]
if z.all() == x_z[15:].all():
    print("Calculated z's from 4.d and 4.f (here) are the same.")
    print("z calculated from 4.d : {}".format(z))
    print("z calculated from 4.f(here): {}".format(x_z[15:]))
```

### g. 

In order to satisfy KKT conditions, C has to be either square or wide. C keeps the additional (constraining) equations, which are continuity equations in our case. The model function (piecewise function) is divided into three parts and the intersection points of those parts (there are two: -0.5, 0.5) brings three extra equations to satisfy continuity. (Function itself, its the first and the second derivatives.) Therefore we have 6 extra equations in C. In order to make C wide, each row has to have at least 6 coefficient. However, since each row has divided into three parts, keeping the coefficients of each function, each part should have bring at least two coefficients and they bring D+1 coefficient since index starts from 0.

Thus, D = 0 makes C matrix tall (6x3) and violates KKT conditions. D can be 1 (makes C square), 2, 3, or 4.
